```python
import numpy as np
```

```python
# Function to compute the classification error
def compute_error(X, y, W, b):
    return np.sum(np.sign(np.dot(X, W) + b) != y)

# Input data (4 samples, 2D features)
X = [(1, 1), (2, -2), (1, -1), (-1, -1)]
y = [1, 1, -1, 1]

# Convert to numpy arrays
X = np.array(X)
y = np.array(y)

# Initialize bias and weight vector
b = 0
num_examples = X.shape[0]
dim_x = X.shape[1]
W = np.zeros(dim_x)

# Control flags and counters
flag = 1               # Indicates if any classification error occurred
iteration = 0
max_iter = 100         # Maximum number of iterations

# Pocket variables to store the best solution seen so far
W_pocket = np.zeros(dim_x)
E_pocket = float('inf')  # Best error so far (start with infinity)
b_pocket = 0

# Perceptron training loop with pocket algorithm
while flag == 1 and iteration < max_iter:
    flag = 0
    iteration += 1
    for i in range(num_examples):
        # Check if the current sample is misclassified
        if y[i] * (np.dot(W, X[i]) + b) <= 0:
            # Update weights and bias
            W += y[i] * X[i]
            b += y[i]
            # Evaluate current error
            E = compute_error(X, y, W, b)
            # Save best model if current one is better
            if E < E_pocket:
                W_pocket = np.copy(W)
                E_pocket = E
                b_pocket = b
            flag = 1  # A change occurred, continue looping
            break     # Only one update per iteration (standard perceptron behavior)

# Output results
print(f"E = {E_pocket}")
print(f"b = {b_pocket}")
print(f"number of iteration = {iteration}")
print(f"W = {W_pocket}")
print(f"hyperplan:   x2 = {-W_pocket[0]/W_pocket[1]}*x1 + {-b_pocket/W_pocket[1]}")
```
